# COMCAST CONSUMER COMPLAINTS

## IMPORTING LIBRARIES





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
#LOADING AND READING THE DATA

a = pd.read_csv('Comcast_telecom_complaints_data.csv')

dataframe = pd.DataFrame(a)

dataframe.head(10)
```

```python
#SORTING THE DATAFRAME ACCORDING TO THE TICKET ID

dataframe.sort_values('Ticket #' , inplace = True)

dataframe
```

```python
#DROPING 'DATE' COLUMN AS IT IS NOT REQUIRED FOR ANY PURPOSE

dataframe.drop(['Date'] , inplace = True, axis = 1)
dataframe
```

```python
# CONVERTING THE 'Date_month_year' COLUMN INTO DATETIME FORMAT

dataframe['Date_month_year'] = pd.to_datetime(dataframe['Date_month_year'])
dataframe
```

### MONTHLY CHART

```python
#EXTRACTING MONTH VALUES 

month_chart = dataframe.groupby(pd.Grouper(key = 'Date_month_year' , freq = 'M'))
month_chart.size().plot(kind = 'line')
```

### DAILY CHART

```python
#EXCTRACTING COMPLAINTS ON DAILY BASIS

daily_chart = dataframe.groupby(pd.Grouper(key = 'Date_month_year' , freq = 'D'))
daily_chart.size().plot(kind = 'line')
```

## FREQUENCY OF COMPLAINT TYPES

```python
dataframe['Customer Complaint'] = dataframe['Customer Complaint'].str.upper()
dataframe['Customer Complaint']
dataframe['Customer Complaint'].str.split(expand=True).stack().value_counts()
```

### FREQUENCY PLOT FOR ALL THE COMPLAINTS

```python
plt.figure(figsize=(10,50))
plt.title('CITY-WISE COMPLAINTS')
sns.countplot(y='Customer Complaint' , data = dataframe)
```

### CREATING A NEW CATEGORICAL COLUMN WITH VALUES AS OPEN AND CLOSED 

```python
dataframe['new_status'] = ['Open' if Status =="Open" or Status =="Pending" else "Closed" for Status in dataframe['Status'] ]
dataframe
```

```python
complaint_status = dataframe.groupby(["State","new_status"]).size().unstack().fillna(0)
complaint_status
```

```python
complaint_status.plot(kind='barh', stacked = True, figsize=(25,25))
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('COMPLAINTS STATUS ACROSS STATES \n' , fontsize = 30)
plt.show()
```

### STATE WISE COMPLAINT COUNT

```python
state_complaint_status = dataframe.groupby(['State']).size()


state_complaint_status = state_complaint_status.sort_values(ascending = False)


state_complaint_status = state_complaint_status.reset_index()


state_complaint_status = state_complaint_status.rename({0 : 'count'},axis = 1)

state_complaint_status
```

### STATE WITH THE MAXIMUM COMPLAINTS

```python
state_complaint_status.max()
```

### PERCENTAGE OF UNRESOLVED COMPLAINTS ACROSS STATES

```python
percentage_complaint = complaint_status['Open'].sort_values(ascending = False)

complaint_status['total complaint %'] =(percentage_complaint/percentage_complaint.sum()) * 100


complaint_status = complaint_status.sort_values(['total complaint %'] , ascending = False)

complaint_status.max()
```
